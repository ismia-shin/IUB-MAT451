```python
# Simulation for Initial value P0 = 50

P0 = 50
Population = []

for i in range(30):
    P_next = P0 + 0.7 * P0 * (1 - P0/500)
    Population.append(P_next)
    P0 = P_next

print(Population)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# The recursive function for the growth model
def logistic_model(P_n):
    R = 0.7
    K = 500
    P_next = P_n + R * P_n * (1 - P_n / K)
    return P_next

# --- Simulation Parameters ---
initial_values = [20, 200, 600]

num_steps = 30
time_steps = np.arange(num_steps + 1)

# --- Trajectory Generation ---
plt.figure(figsize=(10, 6))
plt.title('Growth Model Trajectories')
plt.xlabel('Time Step (n)')
plt.ylabel('Population (P_n)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(y=500, color='#f0607d', linestyle=':', linewidth=1, label='Carrying Capacity (K=500)')


for P0 in initial_values:
    P_trajectory = np.zeros(num_steps + 1)
    P_trajectory[0] = P0

    for n in range(num_steps):
        P_trajectory[n+1] = logistic_model(P_trajectory[n])

    plt.plot(time_steps, P_trajectory, marker='.', linestyle='-', linewidth=2, label=f'P = {P0}')

# --- Plot Customization ---
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
